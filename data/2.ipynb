```python
import json

with open("result.json", "r") as read_file:
    data = json.load(read_file)
```

```python
import vk_api

vk_session = vk_api.VkApi(token="<SOME_TOKEN>")

vk = vk_session.get_api()

res = vk.groups.getById(group_ids=[x["id"] for x in list(data["nodes"])])
```

```python
for i in range(len(data["nodes"])):
    data["nodes"][i]["name"] = res[i]["name"]
    data["nodes"][i]["pic"] = res[i]["photo_100"]
```

```python
import os
import requests

path = "imgs"

try:
    os.mkdir(path)
except OSError:
    print ("Creation of the directory %s failed" % path)
else:
    print ("Successfully created the directory %s " % path)

os.chdir(path)

for i in range(len(data["nodes"])):
    response = requests.get(data["nodes"][i]["pic"])
    file = open(str(data["nodes"][i]["id"])+".jpg", "wb")
    file.write(response.content)
    file.close()
```

```python
for i in range(len(data["nodes"])):
    data["nodes"][i] = {"id":data["nodes"][i]["id"], "name":data["nodes"][i]["name"]}
```
